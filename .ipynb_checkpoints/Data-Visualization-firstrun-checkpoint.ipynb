With this part of the project I am attempting to compose a multi-layer, interactive data visualization which allows for the usage of several different layers so more and less detail can be shown. 

After looking at the various options, I decided on using the Open Source GeoViews python library. I had also looked at other python libraries as well as the javascript D3 one, but decided on GeoViews for its' simplicity and elegance.

First I will go through and plot a basic interactive map so I understand how the library works.

```python
import numpy as np
import xarray as xr
import pandas as pd
import holoviews as hv
import geoviews as gv
import iris
import cartopy

from cartopy import crs
from cartopy import feature as cf
from geoviews import feature as gf

hv.notebook_extension('bokeh','matplotlib')
%output backend='matplotlib'
%opts Feature [projection=crs.Robinson()]
```

Here is an example of how features can be combined.

```python
gf.coastline + gf.ocean + gf.ocean*gf.land*gf.coastline
```

```python
%%opts Feature.Lines (facecolor='none' edgecolor='gray')
graticules = gv.Feature(cf.NaturalEarthFeature(category='physical', name='graticules_30',scale='110m'), group='Lines')
graticules
```

Here is an example of how overlay of different features works. 

```python
%%output size=450
features = hv.Overlay([gf.ocean, gf.land, graticules, gf.rivers, gf.lakes, gf.borders, gf.coastline])
features
```

Here is an example utilizing the bokeh library rather than the matplotlib. This is an example of how the interactivity of the Bokeh maps work.

```python
%%output backend='bokeh'
%%opts Overlay [width=600 height=500 xaxis=None yaxis=None] Feature.Lines (line_color='gray' line_width=0.5)
features
```
